## Diffrential Equations

$$\frac{dS}{dt} = -\beta SI,$$
$$\frac{dI}{dt} = \beta SI - \gamma I,$$
$$\frac{dR}{dt} = \gamma I.$$

### Model Variables

- **S(t)**: Number of susceptible individuals  
- **I(t)**: Number of infectious individuals  
- **R(t)**: Number of removed individuals (recovered and immune, or deceased)

### Model Parameters

- **β (beta)**: Infection rate constant  
- **γ (gamma)**: Recovery rate constant


Let

$$
\mathbf{V}(t) =
\begin{bmatrix}
S(t) \\
I(t) \\
R(t)
\end{bmatrix}
$$

denote the state vector of the system. The dynamics of the model can then be written as

$$
\frac{d\mathbf{V}}{dt}
= \mathbf{f}\bigl( t, \mathbf{V} \bigr)
= \mathbf{f}\bigl(t, S(t), I(t), R(t)\bigr)
= \begin{bmatrix}
\frac{dS}{dt} \\
\frac{dI}{dt} \\
\frac{dR}{dt}
\end{bmatrix}.
$$


```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import odeint
from matplotlib.animation import FuncAnimation
from IPython.display import HTML
```

```python
def dVdt(V, t, beta, gamma):
    S, I, R = V
    dSdt = -beta * S * I
    dIdt = beta * S * I - gamma * I
    dRdt = gamma * I
    return [dSdt, dIdt, dRdt]
```

```python
S0 = 0.99  # Initial susceptible population
I0 = 0.01  # Initial infected population
R0 = 0.0   # Initial recovered population
V0 = (S0, I0, R0)

beta = 0.3
gamma = 0.1
```

```python
t = np.linspace(0, 100, 1000)
sol = odeint(dVdt, y0 = V0, t=t, args=(beta, gamma))
```

```python
sol.shape
```

```python
plt.figure(figsize=(7, 4), dpi=300)
plt.plot(t, sol[:,0], label='S')
plt.plot(t, sol[:,1], label='I')
plt.plot(t, sol[:,2], label='R')
plt.xlabel('t')
plt.ylabel('N(t)')
plt.title('Logistic growth model')
plt.grid(ls='--')
plt.legend()
plt.tight_layout()
plt.show()
```

## Stronger recovery rate
